```python
from urllib.parse import urlparse
import re

def getDomain(url):
    # Ensure the URL has a scheme (http or https)
    if not url.startswith(('http://', 'https://')):
        url = 'http://' + url  # Add http as a default scheme

    domain = urlparse(url).netloc
    if re.match(r"^www\.", domain):  # Use escaped dot for regex
        domain = domain.replace("www.", "")
    return domain


```

```python
import whois
from datetime import datetime
from urllib.parse import urlparse
import re

def get_domain_age(url):
    domain=getDomain(url)
    try:
        domain_info = whois.whois(domain)
        creation_date = domain_info.creation_date
        if isinstance(creation_date, list):  # Some domains have multiple creation dates; use the first
            creation_date = creation_date[0]

        # Check if creation_date is valid
        if creation_date is None:
            age=-1

        # Calculate age in years
        age_in_months = (datetime.now() - creation_date).days // 30  # Convert days to years
        age= age_in_months
    except Exception as e:
        age=-1
# Check domain age and print 0 or 1 based on the age condition
    if age!=-1 and age > 6:
        return 0
    else:
        return 1
```

```python
import whois
from datetime import datetime

def get_domain_end_period(url):
    domain=getDomain(url)
    try:
        domain_info = whois.whois(domain)
        expiration_date = domain_info.expiration_date
        if isinstance(expiration_date, list):  # Some domains have multiple expiration dates; use the first
            expiration_date = expiration_date[0]

        # Check if expiration_date is valid
        if expiration_date is None:
            end_period=-1

        # Calculate remaining time in days
        remaining_days = (expiration_date - datetime.now()).days

        # Convert days to months
        remaining_months = remaining_days // 30
        end_period=remaining_months
    except Exception as e:
        end_period=-1
    print(end_period)
    # Check end period and print 1 or 0 based on the condition
    if end_period < 6:
        return 1  # Phishing: End period is lesser than 6 months
    else:
        return 0  # Legitimate: End period is 6 months or more

```

```python
import whois
import socket

def check_dns_records(domain):
    try:
        # Try to resolve the domain using DNS
        socket.gethostbyname(domain)
        return True  # DNS record exists
    except socket.gaierror:
        return False  # No DNS record found

def check_whois_data(domain):
    try:
        # Get WHOIS information
        domain_info = whois.whois(domain)

        # Check if WHOIS records are empty or invalid
        if domain_info.creation_date is None or domain_info.registrar is None:
            return False  # WHOIS data is incomplete
        return True  # WHOIS data exists and is valid
    except Exception as e:
        return False  # No WHOIS data available

def is_phishing(url):
    domain=getDomain(url)
    # Check both DNS and WHOIS data to determine if the domain is phishing
    has_dns = check_dns_records(domain)
    has_whois = check_whois_data(domain)

    if not has_dns or not has_whois:
        return 1  # Phishing: Missing DNS or WHOIS data
    return 0  # Legitimate: DNS and WHOIS data found

```

```python
import requests
from bs4 import BeautifulSoup

def check_empty_title(url):
    try:
        # Send HTTP GET request to fetch the page content
        response = requests.get(url)

        # Check if the response status code is 200 (OK)
        if response.status_code == 200:
            # Parse the page content using BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')

            # Find the <title> tag
            title_tag = soup.find('title')

            # Check if the title tag exists and if it's empty
            if title_tag is None or not title_tag.text.strip():
                return 1  # Phishing site (empty title)
            else:
                return 0  # Legitimate site (non-empty title)
        else:
            return 1  # Phishing or error in fetching content

    except Exception as e:
        return 1  # Phishing or error in processing

# Example usage

```

```python
import tldextract

def tld_in_subdomain(url):
    # Extract the domain components using tldextract
    ext = tldextract.extract(url)

    # ext.subdomain: the subdomain part (if any)
    # ext.domain: the domain name (example for "example.com" it returns "example")
    # ext.suffix: the top-level domain (TLD) part (example for "example.com" it returns "com")

    subdomain = ext.subdomain
    tld = ext.suffix

    # Check if the TLD is in the subdomain
    if tld in subdomain:
        return 1  # Return True and the extracted components (phishing)
    else:
        return 0  # Return False and the extracted components (legitimate)

```

```python
import requests
from bs4 import BeautifulSoup

def is_indexed(url):
    # Construct the Google search query with 'site:' to check if URL is indexed
    query = f"https://www.google.com/search?q=site:{url}"

    # Set headers to mimic a real browser request
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36"
    }

    response = requests.get(query, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    # Check if any search results are returned for the URL
    if "did not match any documents" in response.text:
        return 1  # Not indexed
    else:
        return 0 # Indexed
```

```python
# 2.Checks for IP address in URL (Have_IP)
def havingIP(url):
  try:
    ipaddress.ip_address(url)
    ip = 1
  except:
    ip = 0
  return ip
```

```python

# 3.Checks the presence of @ in URL (Have_At)
def haveAtSign(url):
  if "@" in url:
    at = 1
  else:
    at = 0
  return at
```

```python
# 4.Finding the length of URL and categorizing (URL_Length)
def getLength(url):
  if len(url) < 54:
    length = 0
  else:
    length = 1
  return length
```

```python

# 5.Gives number of '/' in URL (URL_Depth)
def getDepth(url):
  s = urlparse(url).path.split('/')
  depth = 0
  for j in range(len(s)):
    if len(s[j]) != 0:
      depth = depth+1
  return depth
```

```python

# 6.Checking for redirection '//' in the url (Redirection)
def redirection(url):
  pos = url.rfind('//')
  if pos > 6:
    if pos > 7:
      return 1
    else:
      return 0
  else:
    return 0
```

```python
# 7.Existence of HTTPS Token in the Domain Part of the URL (https_Domain)
def httpDomain(url):
  domain = urlparse(url).netloc
  if 'http' in domain:
    return 1
  else:
    return 0
```

```python
def httpsDomain(url):
  domain = urlparse(url).netloc
  if 'https' in domain:
    return 1
  else:
    return 0
```

```python


#listing shortening services
shortening_services = r"bit\.ly|goo\.gl|shorte\.st|go2l\.ink|x\.co|ow\.ly|t\.co|tinyurl|tr\.im|is\.gd|cli\.gs|" \
                      r"yfrog\.com|migre\.me|ff\.im|tiny\.cc|url4\.eu|twit\.ac|su\.pr|twurl\.nl|snipurl\.com|" \
                      r"short\.to|BudURL\.com|ping\.fm|post\.ly|Just\.as|bkite\.com|snipr\.com|fic\.kr|loopt\.us|" \
                      r"doiop\.com|short\.ie|kl\.am|wp\.me|rubyurl\.com|om\.ly|to\.ly|bit\.do|t\.co|lnkd\.in|db\.tt|" \
                      r"qr\.ae|adf\.ly|goo\.gl|bitly\.com|cur\.lv|tinyurl\.com|ow\.ly|bit\.ly|ity\.im|q\.gs|is\.gd|" \
                      r"po\.st|bc\.vc|twitthis\.com|u\.to|j\.mp|buzurl\.com|cutt\.us|u\.bb|yourls\.org|x\.co|" \
                      r"prettylinkpro\.com|scrnch\.me|filoops\.info|vzturl\.com|qr\.net|1url\.com|tweez\.me|v\.gd|" \
                      r"tr\.im|link\.zip\.net"


# 8. Checking for Shortening Services in URL (Tiny_URL)
def tinyURL(url):
    match=re.search(shortening_services,url)
    if match:
        return 1
    else:
        return 0

```

```python

# 9.Checking for Prefix or Suffix Separated by (-) in the Domain (Prefix/Suffix)
def prefixSuffix(url):
    if '-' in urlparse(url).netloc:
        return 1            # phishing
    else:
        return 0            # legitimate
```

```python


# 15. IFrame Redirection (iFrame)
def iframe(response):
  if response == "":
      return 1
  else:
      if re.findall(r"[|]", response.text):
          return 0
      else:
          return 1

```

```python
# 16.Checks the effect of mouse over on status bar (Mouse_Over)
def mouseOver(response):
  if response == "" :
    return 1
  else:
    if re.findall("", response.text):
      return 1
    else:
      return 0
```

```python
# 17.Checks the status of the right click attribute (Right_Click)
def rightClick(response):
  if response == "":
    return 1
  else:
    if re.findall(r"event.button ?== ?2", response.text):
      return 0
    else:
      return 1
```

```python

# 18.Checks the number of forwardings (Web_Forwards)
def forwarding(response):
  if response == "":
    return 1
  else:
    if len(response.history) <= 2:
      return 0
    else:
      return 1
```

```python
#Function to extract features
def featureExtraction(url,label):

  features = []
  #Address bar based features (10)
  features.append(getDomain(url))
  features.append(havingIP(url))
  features.append(haveAtSign(url))
  features.append(getLength(url))
  features.append(getDepth(url))
  features.append(redirection(url))
  features.append(httpDomain(url))
  features.append(httpsDomain(url))
  features.append(tinyURL(url))
  features.append(prefixSuffix(url))

  features.append(get_domain_age(url))
  features.append(get_domain_end_period(url))
  features.append(is_phishing(url))
  features.append(check_empty_title(url))
  features.append(tld_in_subdomain(url))
  features.append(is_indexed(url))


  # HTML & Javascript based features (4)
  try:
    response = requests.get(url)
  except:
    response = ""
  features.append(iframe(response))
  features.append(mouseOver(response))
  features.append(rightClick(response))
  features.append(forwarding(response))
  features.append(label)

  return features
```
